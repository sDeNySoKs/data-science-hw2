```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_survey = pd.read_csv("2017_jun_final - 2017_jun_final.csv")
df_survey.head()
```

```python
df_survey.shape
```

```python
df_survey.dtypes
```

```python
missing = df_survey.isnull().sum() / len(df_survey)
print(missing)
```

```python
cols_with_nans = df_survey.columns[df_survey.isnull().any()].tolist()
cols_with_nans.remove('Мова програмування')
df_survey = df_survey.drop(cols_with_nans, axis=1)
print(df_survey)
```

```python
missing = df_survey.isnull().sum() / len(df_survey)
print(missing)
```

```python
df_survey = df_survey.dropna()
```

```python
df_survey.shape
```

```python
python_data = df_survey[df_survey['Мова програмування'] == 'Python']
print(python_data)
```

```python
python_data.shape
```

```python
salary_by_position = python_data.groupby('Посада').agg({'Зарплата на місяць': ['min', 'max']})
print(salary_by_position)
```

```python
def fill_avg_salary(row):
    return (row[('Зарплата на місяць', 'min')] + row[('Зарплата на місяць'), 'max']) / 2

salary_by_position['avg'] = salary_by_position.apply(fill_avg_salary, axis=1)
print(salary_by_position)
```

```python
print(salary_by_position['avg'].describe())
```

```python
salary_by_position.to_csv("python_salaries_report.csv")
```

```python
df_survey['Мова програмування'].value_counts().head(10).plot(kind='bar', title='Топ-10 мов програмування')
plt.show()

df_survey['Зарплата на місяць'].plot(kind='hist', title='Розподіл зарплат айтішників', bins=20)
plt.show()

salary_by_position['avg'].plot(kind='bar', title='Середня зарплата Python-розробників', color='green')
plt.show()
```
